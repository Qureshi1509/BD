```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load your data (adjust file paths as necessary)
hero_network_df = pd.read_csv("C:/Users/Qureshi/Downloads/hero-network.csv")

# Create a graph from the hero network
G = nx.Graph()
for _, row in hero_network_df.iterrows():
    G.add_edge(row['hero1'], row['hero2'])

# Select a random subset of nodes (e.g., 500 nodes) for approximation
subset_nodes = random.sample(G.nodes(), 10)

# Calculate approximate betweenness centrality using the subset
betweenness_centrality_approx = nx.betweenness_centrality_subset(G, sources=subset_nodes, targets=subset_nodes)

# Display the top 5 nodes by betweenness centrality (approximate)
top_betweenness_centrality_approx = sorted(betweenness_centrality_approx.items(), key=lambda x: x[1], reverse=True)[:5]

print("Top 5 nodes by Betweenness Centrality (Approx):")
for node, centrality in top_betweenness_centrality_approx:
    print(f"{node}: {centrality:.4f}")
```

```python
subset_nodes2 = random.sample(G.nodes(), 250)
subgraph = G.subgraph(subset_nodes2)
# Calculate closeness centrality for all nodes in the graph
closeness_centrality = nx.closeness_centrality(subgraph)

# Display the top 5 nodes by closeness centrality
top_closeness_centrality = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]

print("Top 5 nodes by Closeness Centrality:")
for node, centrality in top_closeness_centrality:
    print(f"{node}: {centrality:.4f}")
```

```python
# Load your data (adjust file paths as necessary)
hero_network_df = pd.read_csv("C:/Users/Qureshi/Downloads/avngr.csv")

# Create a graph from the hero network
G = nx.Graph()
for _, row in hero_network_df.iterrows():
    G.add_edge(row['hero1'], row['hero2'])
# Visualize the network
#plt.figure(figsize=(12, 12))  # Set the figure size for better visibility

# Generate a layout for the graph nodes
pos = nx.spring_layout(G, k=0.9)  # k controls the distance between nodes

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=50, node_color='red', alpha=0.9)
nx.draw_networkx_edges(G, pos, width=0.8, alpha=0.9)

# Optionally, label the nodes
nx.draw_networkx_labels(G, pos, font_size=8, font_color='black', font_family="times new roman")

# Hide axis for better aesthetics
plt.axis('on')

# Show the plot
plt.show()
```
